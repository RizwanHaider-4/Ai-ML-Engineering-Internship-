```python
# Load the Iris dataset using seaborn (already in CSV format)
df = sb.load_dataset("iris")

# Display first 5 rows to inspect the dataset
df.head()
```

```python
# Scatter plot: Sepal length vs Sepal width
sb.scatterplot(
    x='sepal_length',
    y='sepal_width',
    hue='species',  # color by species
    data=df,
    palette='Set2'
)

plt.title("Sepal Length vs Sepal Width by Species")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

```python
# Plot histogram for all numerical features
df.hist(bins=15, figsize=(10,6), color='violet')
plt.suptitle("Feature Distributions")
plt.show()
```

```python
# Box plot to identify outliers for petal width by species
sb.boxplot(x='species', y='petal_width', data=df, palette='Set3')
plt.title("Petal Width by Species (Outlier Detection)")
plt.show()
```
